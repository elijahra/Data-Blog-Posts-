# What are Primative Pythagorean Triples ?

Before we dive into finding a closed form or fomula for all the primative pythagorean tripples we are going to have to define some terms that will allows to create functions to filter our pythagorean triples from other integers.

$\textbf{Pythagorean Triples}$ are the solutions to the pythagorean theorem $a^2 + b^2 = c^2$ for which $\textbf{a}$, $\textbf{b}$, and $\textbf{c}$ are integers  
(postive and negative whole numbers such as $\dots,-2,-1,0,1,2,\dots$). 

 The $\textbf{greatest common divisor}$ or $GCD$ is defined as to a number that divides all three nuimbers $\textbf{a}$, $\textbf{b}$ and, $\textbf{c}$ such that the division of the number remains an integer. The numbers a,b, and c are called $\textbf{relativly prime}$ if the greatest common divisor of the three numbers is $1$.

A $\textbf{Primative Pythagorean Triple}$ or $\textbf{PPT}$ are three relatively prime integers $\textbf{a}$, $\textbf{b}$, and $\textbf{c}$ that satisfy $a^2 + b^2 = c^2$

# Finding a closed form for Pythagorean Triples 

Every point plotted can be written as  $(x,y,z)$ since these are to be pythagorean triples we know that x,y and z satisfy the equation:

$$x^2+y^2=z^2$$

then doing some algebraic manipulation find that 

$$x ^2  = z^2 -y^2$$


$$1  = \left(\frac{z}{x}\right)^2 -\left(\frac{y}{x}\right)^2$$


Factoring we get 


$$ 1 = \left(\frac{z}{x} -\frac{y}{x}\right)\left(\frac{z}{x} + \frac{y}{x}\right)$$

Since the factors mutiply to equal 1 they must be multiplicative inverses of each other the we then assign each of these factors to a ratio


$$\left(\frac{z}{x}+\frac{y}{x}\right) = \frac{A}{B}$$


$$\left(\frac{z}{x} -\frac{y}{x}\right) = \frac{B}{A}$$


Solving those two equations for $\dfrac{z}{x}$ and $\dfrac{y}{x} $

$$\dfrac{z}{x}  = \frac{1}{2}\left(\frac{A}{B} - \frac{B}{A}\right) = \frac{A^2-B^2}{2AB}$$


$$\dfrac{y}{x}  = \frac{1}{2}\left(\frac{A}{B} + \frac{B}{A}\right) = \frac{A^2+B^2}{2AB}$$


We conclude that $x = 2AB$, $y = A^2+B^2$, $z = A^2 -B^2$

$$(x,y,z) = (2AB, A^2+B^2, A^2-B^2)$$
