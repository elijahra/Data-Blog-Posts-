```python
data = loadtxt ("co2_mm_mlo.txt",skiprows = 72)
```

```python
figure(figsize=(12,8))
years = data[:,2]
co2 = data[:,4]
plot(years, co2,"b")
title ("Years v. Co2")
xlabel('Years')
ylabel('Co2')
#legend()
```

## Exercise 2

```python
xbar = mean(years)
ybar = mean(co2)
x2bar = mean(years**2)
x3bar = mean (years**3)
x4bar = mean (years**4)
xybar = mean(years * co2)
x2ybar = mean ((years**2) * co2)
```

```python
A = array([[1,xbar],[xbar, x2bar]])
b = array([ybar, xybar])
```

```python
w0,w1 = linalg.solve(A,b)
linreg = w0 + w1*years
```

***Linear Fit***
------------------

```python
figure(figsize=(12,8))
years = data[:,2]
co2 = data[:,4]
plot(years, co2,"m")
plot (years, linreg, "r")
title("Linear Fit")
xlabel('Years')
ylabel('Co2')
#legend()
```

**Linear Prediction**

```python
linreg2050 = w0 + w1*2050
linreg2100 = w0 + w1*2100
```

```python
print (linreg2050,"micromol/mol")
print (linreg2100,"micromol/mol")
```

**Residual(Linear) Plot**
-------------------

```python
linres = co2 - linreg
```

```python
figure(figsize=(12,8))
title("Linear Residual Plot")
plot (years, linres, "r")
xlabel('Years')
ylabel('Co2')
```

## Findings

When our plot is fitted with our linreg; besides the seasonal signal we have a parabolic curve. This implies that the linear fit is not that acurate because our residual plot is formed from  (data - linreg) since the residual is parabloic there must be a quadratic term left in the funtion in the (years v. Co2) function. So, so then a futher fit is necessary. 

## Exercise 3

**Quadratic Fit**
--------------------

```python
B = array([[1,xbar,x2bar],[xbar,x2bar,x3bar],[x2bar,x3bar,x4bar]])
n = array([ybar,xybar,x2ybar])
```

```python
v0,v1,v2= linalg.solve(B,n)
```

```python
quadreg = v0 + (v1*years)+(v2*(years**2))
```

```python
figure(figsize=(12,8))
years = data[:,2]
co2 = data[:,4]
plot(years, co2,"m")
plot (years, quadreg, "r")
title("Quadratic Fit")
xlabel('Years')
ylabel('Linear Residuals')
#legend()
```

**Quadratic Prediction**

```python
quadreg2050 = v0 + (v1*2050)+(v2*(2050**2))
quadreg2100 = v0 + (v1*2100)+(v2*(2100**2))
```

```python
print (quadreg2050,"micromol/mol")
```

```python
print (quadreg2100,"micromol/mol")
```

**Quadratic Residual Plot**
-------------------------------

```python
quadres = co2 - quadreg
```

```python
figure(figsize=(12,8))
title("Quadratic Residual Plot")
plot(years, quadres )
```

## Findings

This residual plot is flatter and besides the seasonal fluctuations
